# Consensus

### Examples
* Leader election
* Failure detection
* Mutual exclusion
* Something else

Imagine we're trying to get everyone to choose 0 or 1. 

__Validity__ If everyone agrees on 0 (or 1), then that's the answer
__Integrity__ Someone must have suggested 0 or 1 before deciding on it
__Non-Triviality__ Some initial state exists that leads to all 0s or all 1s

### System Model
* Asynchronous > Synchronous, in terms of complexity
    * No message delay bounds
    * No bounds on clock drift rates
    * No max time per process setp

## Consensus in Synchronous Systems
Imagine every process in your have tracks a "value"
Every "round", union your "values" together
If you have 4 processes crash, it will take you a fifth round to have uniform "values" in all processes.
Boom.

## Paxos
Fact: Impossible to distinguish a failed process from an extremely slow message

Paxos provides __safety__ and __eventual liveness__
* Safety: If you reach consensus, every process will have reached the same value
* Eventual liveness: You'll probably reach consensus, if failures & message drop & delays don't get too much in the way

Paxos:
* Uses rounds
* But you advance your round when you hear a message from a future round
* Uses pessimistic timeouts
* The rounds are phased
    * Elect: Elect a leader
    * Bill: Leader proposes a value
    * Law: Leader multicasts the final value

__Election__
1. Choose a higher ballot ID than ever before
1. Multicast your ballot ID
    1. If a value had been decided in the last round, include it too.
1. Processes wait for multicasts, responds "Elect!" to process with highest ballot ID they see.
1. Winner of quorum becomes leader

__Bill__
1. Leader proposes new value over multicast
    1. Or previous value
1. Recipients respond OK and log
    1. Point of No Return: this value will get re-used if the leader should fail before the Law phase.

__Law__
1. If leader gets quorum of OKs to value, multicast this value
1. Processes record published value

__Liveness, or What Could Go Wrong?__
* Process failure.
    * On restart, read log
* Leader fails
    * Round times out
* Messages get dropped
    * Quorums can't be reached - round timeout
* Protocol may never end 
    * Damn.
    * Seriously, you just have to hope.
