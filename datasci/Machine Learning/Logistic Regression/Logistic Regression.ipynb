## Questions
> How does lambda affect under/overfitting?

Underfitting occurs when your model is too simple - not enough features, or to basic of an equation - such as if you're using a linear equation to model a non-linear system.

Overfitting occurs when you've over-matched your model to the training data. Too much feature engineering, such as adding in lots of polynomial terms ($x_{1}^{2}x_{2}^{3}x_{4}^{4}...$), can lead to this. Now we've forced our model to account for every variation in the training data, and we'll be fragile to any new values.

> Why do we exclude the bias term from regularization?

From Tom Mosher, a class Mentor:
Penalizing $\theta_{0}$ results in a system that works badly for both the training set and for new examples, as it starts to shift the system bias toward zero.
